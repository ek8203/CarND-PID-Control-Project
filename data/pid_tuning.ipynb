# PID optimization

```python
import numpy as np
import plotly.offline as py
from plotly.graph_objs import *
import pandas as pd
import math
py.init_notebook_mode()

my_cols=['cte', 'angle','output','speed']
with open('pid_output.txt') as f:
    table_pid_output = pd.read_table(f, sep='\t', header=None, names=my_cols, lineterminator='\n')
    

import plotly.offline as py
from plotly.graph_objs import *

n_step = np.arange(1000)

trace1 = Scatter(
    x=n_step,
    y=table_pid_output['cte'],
    xaxis='x2',
    yaxis='y2',
    name='PID - CTE',
    mode = 'markers'
)

trace2 = Scatter(
    x=n_step,
    y=table_pid_output['angle'],
    xaxis='x2',
    yaxis='y2',
    name = 'Angle' ,
    mode = 'markers'
)

trace3 = Scatter(
    x=n_step,
    y=table_pid_output['output'],
    xaxis='x2',
    yaxis='y2',
    name='Steering Value'
)

trace4 = Scatter(
    x=n_step,
    y=table_pid_output['speed'],
    xaxis='x2',
    yaxis='y2',
    name='Trottle'
)

#data = [trace1, trace2, trace3]
data = [trace1, trace2]
#data = [trace4]

layout = Layout(
    xaxis2=dict(
   
        anchor='x2',
        title='px'
    ),
    yaxis2=dict(
    
        anchor='y2',
        title='py'
    )
)

fig = Figure(data=data, layout=layout)
py.iplot(fig, filename='PID')
```

```python
# PID tuning using Ziegler-Nichols method
Ku = 0.1
ma = [20., 47., 78.]

sum_per = 0.
for i in range(len(ma) - 1):
    sum_per += ma[i+1]-ma[i]

Pu = sum_per/(len(ma)-1)
Kp = 0.6*Ku
Ki = 2. * Kp / Pu
Kd = Kp*Pu/8.

print("Pu = {}\nKp = {}\nKi = {}\nKd = {}".format(Pu, Kp, Ki, Kd))
```
